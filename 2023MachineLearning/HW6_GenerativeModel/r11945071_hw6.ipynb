# Training model use StyleGAN2
--data: 指定用於訓練 GAN 模型的數據集的路徑。\
--name: 指定訓練過程中的模型名稱。\
--results_dir: 指定生成圖像的輸出目錄。\
--models_dir: 指定存儲訓練過程中模型檔案的目錄。\
--network-capacity: 指定生成器和鑑別器的網絡容量，通常用於控制生成圖像的解析度和細節。\
--batch_size: 指定每個訓練批次的大小。\
--num-train-steps: 指定訓練過程中的總步數。\
--image_size: 指定生成圖像的尺寸。\
--load-from: 指定從預訓練模型或者特定步數的模型中加載權重。\
--num_generate: 指定生成圖像的數量。\
--num_image_tiles: 指定在生成圖像時在每個輸出圖像中的瓦片數量。\
reference:chatGPT、stylegan_pytorch
https://github.com/rosinality/stylegan2-pytorch

# Use StyleGAN2 model to generate 1000 photos

```python
!stylegan2_pytorch \
--generate \
--name StyleGAN2 \
--results_dir /kaggle/working/outcome \
--models_dir  /kaggle/working \
--network-capacity 70 \
--batch_size 50 \
--image_size 256 \
--load-from {30} \
--num_generate 1000 \
--num_image_tiles 1
```

```python
import os
import shutil
path = "/kaggle/working/outcome/StyleGAN2"
new_path = "/kaggle/working/result/"
#創建資料夾
if os.path.isdir(new_path) == False:
    os.mkdir(new_path)

for img in os.listdir(path):
    if img[-7:] == 'ema.jpg':
        shutil.move(os.path.join(path,img), os.path.join(new_path,img))
for idx, img in enumerate(os.listdir(new_path)):
    os.rename(os.path.join(new_path, img), os.path.join(new_path, str(idx+1)+'.jpg'))
```
